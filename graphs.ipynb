```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("/Users/sarabjotsingh/Downloads/filtered_dataset.csv")
```

```python
# Histogram of Odometer Readings
plt.figure(figsize=(8, 6))
sns.histplot(data['odometer'], bins=20, kde=True)
plt.title('Histogram of Odometer Readings')
plt.xlabel('Odometer Reading')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart of Selling Prices by Make
plt.figure(figsize=(10, 6))
sns.barplot(x='make', y='sellingprice', data=data)
plt.title('Average Selling Prices by Make')
plt.xlabel('Make')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of Selling Prices by Body Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='body', y='sellingprice', data=data)
plt.title('Selling Prices by Body Type')
plt.xlabel('Body Type')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Pie Chart of Car Colors
plt.figure(figsize=(8, 8))
data['color'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Car Colors')
plt.ylabel('')
plt.show()
```

```python
# Time Series Line Plot of Selling Prices over Time
data['saledate'] = pd.to_datetime(data['saledate'])
plt.figure(figsize=(10, 6))
sns.lineplot(x='saledate', y='sellingprice', data=data)
plt.title('Selling Prices over Time')
plt.xlabel('Sale Date')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Chart of Average Selling Prices by Trim
plt.figure(figsize=(10, 6))
sns.barplot(x='trim', y='sellingprice', data=data)
plt.title('Average Selling Prices by Trim')
plt.xlabel('Trim')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Pairplot of Numerical Features
sns.pairplot(data[['odometer', 'sellingprice']])
plt.suptitle('Pairplot of Numerical Features')
plt.show()
```

```python
# Box Plot of Selling Prices by State
plt.figure(figsize=(10, 6))
sns.boxplot(x='state', y='sellingprice', data=data)
plt.title('Selling Prices by State')
plt.xlabel('State')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Violin Plot of Selling Prices by Body Type and Transmission
plt.figure(figsize=(10, 6))
sns.violinplot(x='body', y='sellingprice', hue='transmission', data=data, split=True)
plt.title('Selling Prices by Body Type and Transmission')
plt.xlabel('Body Type')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Stacked Bar Chart of Number of Listings by Make and Body Type
listings_by_make_body = data.groupby(['make', 'body']).size().unstack()
listings_by_make_body.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Number of Listings by Make and Body Type')
plt.xlabel('Make')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.legend(title='Body Type')
plt.show()
```

```python
# Pie Chart of Transmission Types
plt.figure(figsize=(8, 8))
data['transmission'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Transmission Types')
plt.ylabel('')
plt.show()
```

```python
# Histogram of Age of Vehicles
current_year = pd.Timestamp.now().year
data['vehicle_age'] = current_year - data['year']
plt.figure(figsize=(10, 6))
sns.histplot(data['vehicle_age'], bins=10, kde=True)
plt.title('Histogram of Vehicle Age')
plt.xlabel('Age of Vehicle')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot of Selling Prices by Interior Color
plt.figure(figsize=(12, 6))
sns.boxplot(x='interior', y='sellingprice', data=data)
plt.title('Selling Prices by Interior Color')
plt.xlabel('Interior Color')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot of Selling Prices by Mileage
plt.figure(figsize=(10, 6))
sns.lineplot(x='odometer', y='sellingprice', data=data)
plt.title('Selling Prices by Mileage')
plt.xlabel('Odometer Reading')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Box Plot of Selling Prices by Trim and Body Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='trim', y='sellingprice', hue='body', data=data)
plt.title('Selling Prices by Trim and Body Type')
plt.xlabel('Trim')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.legend(title='Body Type')
plt.show()
```

```python
# Facet Grid of Selling Prices by State
g = sns.FacetGrid(data, col='state', col_wrap=4, height=4, aspect=1.2)
g.map(sns.histplot, 'sellingprice')
g.set_titles('{col_name}')
plt.subplots_adjust(top=0.9)
plt.suptitle('Distribution of Selling Prices by State')
plt.show()
```

```python
# Box Plot of Selling Prices by Year and Make
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='sellingprice', hue='make', data=data)
plt.title('Selling Prices by Year and Make')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.legend(title='Make')
plt.show()
```

```python
# Count the occurrences of each body type
body_type_counts = data['body'].value_counts()

# Sort body types by count
sorted_body_types = body_type_counts.sort_values()

# Plot horizontal bar chart
plt.figure(figsize=(10, len(sorted_body_types) * 0.15))  # Adjust figure height based on the number of body types
plt.barh(sorted_body_types.index, sorted_body_types.values, color='skyblue')  # You can change the color if desired
plt.title('Count of Most Common Body Types')
plt.xlabel('Count')
plt.ylabel('Body Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines for better organization
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Count the occurrences of each interior type
interior_counts = data['interior'].value_counts()

# Define the explode parameter (explode the second segment)
explode = [0 if i != 1 else 0.1 for i in range(len(interior_counts))]

# Plot pie chart without labels and percentages
plt.figure(figsize=(8, 8))
plt.pie(interior_counts, explode=explode, autopct='', startangle=140, labels=None)

# Add legend
plt.legend(interior_counts.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Interior Types')
plt.show()
```
